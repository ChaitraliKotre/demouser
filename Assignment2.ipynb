```python
print("14 Akash Choudhary\n")

# Import the load_wine function from sklearn.datasets module
from sklearn.datasets import load_wine

# Load the wine dataset using the load_wine() function
wine_data = load_wine()

# Access the data attribute of the wine dataset to retrieve the feature matrix
# The feature matrix contains the data points, where each row represents a sample (wine) and each column represents a feature
# Features might include alcohol content, malic acid, ash, etc.
wine_features = wine_data.data

# Print the feature matrix
print(wine_features)
```

```python
print("14 Akash Choudhary\n")

# Create a DataFrame from the wine dataset features, using feature names as column headers
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# Add a new column to the DataFrame for the target labels
# The target labels represent the class of each sample (wine) in the dataset
# These labels are added to the DataFrame under the column name "target"
wine_df["target"] = wine_data.target

# Display the first few rows of the DataFrame
print(wine_df.head())
```

```python
print("14 Akash Choudhary\n")

# Display concise summary information about the DataFrame
wine_df.info()
```

```python
print("14 Akash Choudhary\n")

# Display descriptive statistics for numerical columns in the DataFrame
wine_df.describe()
```

```python
print("14 Akash Choudhary\n")

# Import StandardScaler from scikit-learn
from sklearn.preprocessing import StandardScaler

# Copy the features (X) and target labels (Y) from the wine DataFrame
X = wine_df[wine_data.feature_names].copy()
Y = wine_df["target"].copy()

# Initialize a StandardScaler object
scaler = StandardScaler()

# Fit the StandardScaler to the features (X)
# This computes the mean and standard deviation of each feature in X
scaler.fit(X)

# Transform the features (X) using the computed mean and standard deviation
X_scaled = scaler.transform(X.values)

# Print the scaled values of the first sample
print(X_scaled[0])
```

```python
print("14 Akash Choudhary\n")

from sklearn.model_selection import train_test_split

# Split the scaled features (X_scaled) and target labels (Y) into training and testing sets
# train_size=7 indicates that 70% of the data will be used for training and 30% for testing
# random_state=25 sets the random seed for reproducibility
x_train_scaled, x_test_scaled, y_train, y_test = train_test_split(X_scaled, Y, train_size=12, random_state=25)

# Print the proportions of the training and testing sets
print(f"Train Size : {round(len(x_train_scaled) / len(X) * 100)} % \nTest size : {round(len(x_test_scaled) / len(X) * 100)} %")
```

```python
print("14 Akash Choudhary\n")

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Instantiate the Logistic Regression, Support Vector Machine, and Decision Tree models
logistic_model = LogisticRegression()
svm_model = SVC()  # SVM = Support Vector Machine
tree_model = DecisionTreeClassifier()

# Train the models on the scaled training data
logistic_model.fit(x_train_scaled, y_train)
svm_model.fit(x_train_scaled, y_train)
tree_model.fit(x_train_scaled, y_train)

# Make predictions using each model on the scaled test data
logistic_preds = logistic_model.predict(x_test_scaled)
svm_preds = svm_model.predict(x_test_scaled)
tree_preds = tree_model.predict(x_test_scaled)

# Print predictions for each model
print("Logistic Regression Predictions:\n", logistic_preds)
print("\nSupport Vector Machine Predictions:\n", svm_preds)
print("\nDecision Tree Predictions:\n", tree_preds)
```

```python
print("14 Akash Choudhary\n")

from sklearn.metrics import classification_report

# Store model predictions in a dictionary
model_preds = {
    "Logistic Regression: ": logistic_preds,
    "Support Vector Machine : ": svm_preds,
    "Tree: ": tree_preds
}

# Iterate over each model's predictions and print the classification report
for model, preds in model_preds.items():
    print(f"{model}Results : \n {classification_report(y_test, preds)}", sep="\n\n")
```
